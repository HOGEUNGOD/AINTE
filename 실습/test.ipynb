```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

raw_file = pd.read_csv('./data/tension.csv', encoding='CP949')
raw_file
```

```python
tension_data = np.array(raw_file)
tension_data
```

```python
strain, stress = tension_data[:,4]*10/25, tension_data[:,3]*3000/18
```

```python
fig, ax = plt.subplots()
ax.plot(strain, stress)
plt.show()
```

```python
stress = savgol_filter(stress, 201, 1)
strain = savgol_filter(strain, 201, 1)
```

```python
fig, ax = plt.subplots()
ax.plot(strain, stress)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(strain[0:1200], stress[0:1200], color='red', ls=':', lw=10)
ax.set_ylabel('Strain')
ax.set_ylabel('Stress')

plt.show()
```
